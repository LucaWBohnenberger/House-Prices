```python
import pandas as pd
```

```python
base = pd.read_csv('train.csv')
```

```python
base.head()
```

```python
base.shape
```

```python
base.info()
```

```python
(base.isnull().sum()/base.shape[0]).sort_values(ascending=False).head(15) > 0.1
```

```python
kill = base.columns[base.isnull().sum()/base.shape[0] > 0.1]
kill
```

```python
base.drop(columns=kill, inplace=True)
```

```python
base
```

```python
usar = base.columns[base.dtypes != 'object']
usar
```

```python
base_util = base.loc[:, usar]
base_util
```

```python
base_util.isnull().sum().sort_values(ascending=False).head(3)
```

```python
base_util.fillna(-1, inplace=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = base_util.drop(columns='SalePrice')
y = base_util['SalePrice']
```

```python
X_treino, X_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.1, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg_rl = LinearRegression().fit(X_treino, y_treino)    
```

```python
y_rl = reg_rl.predict(X_teste)
```

```python
from sklearn import tree
```

```python
reg_ar = tree.DecisionTreeRegressor(random_state=17).fit(X_treino, y_treino)
```

```python
y_ar = reg_ar.predict(X_teste)
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
reg_kn = KNeighborsRegressor(n_neighbors=4).fit(X_treino, y_treino)
```

```python
y_knn = reg_kn.predict(X_teste)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
print(mean_absolute_error(y_teste, y_rl))
print(mean_squared_error(y_teste, y_rl))
```

```python
print(mean_absolute_error(y_teste, y_ar))
print(mean_squared_error(y_teste, y_ar))
```

```python
print(mean_absolute_error(y_teste, y_knn))
print(mean_squared_error(y_teste, y_knn))
```
